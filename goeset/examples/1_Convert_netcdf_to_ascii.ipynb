# Convert NetCDF File to ASCII

This jupyter notebook deomonstrates the `netcdf2ascii.py` utility that converts a GOES ET NetCDF file to the legacy ascii format.

**NOTE:** the input NetCDF file `data\fl.et.2019.v.0.1.nc` is not provided in this repository. Please download the file separately and place in the `data\` directory or update the code block below to reflect the location of the file on your system.

Run the `netcdf2sacii.py` script with the required argument specifying the input file name. User may optionally supply the name of the output file, otherwise a name is generated automatically. This generally will take a few minutes to run due to the large size of the output file to be written.

```python
%run ..\netcdf2ascii.py data\fl.et.2019.v.1.0.nc
```

Lastly, let's take a look at the output file as a `Pandas.DataFrame` object. We'll just read in the first couple of days of data so this runs quickly (loading the entire dataset takes several minutes depending on disk i\o speed).

```python
# import the library
import goeset

fname = r'data\fl.et.2019.v.1.0.txt'
etf = goeset.GoesAsciiFile(fname)

# Fetch only the first 2 days of data
npix = 39175
nday = 2
df = etf.get_dataframe(nrows=npix * nday)
df
```
